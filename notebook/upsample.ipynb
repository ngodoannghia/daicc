```python
import pandas as pd
```

```python
data_train = pd.read_csv('../analysis/train.csv')
label_train = pd.read_csv('../analysis/train_label.csv')
```

```python
data_train.head(2)
```

```python
label_train.head(2)
```

```python
df_label_anomalous_1 = label_train[label_train["anomalous"] == 1]
```

```python
block_id_anomalous_1 = df_label_anomalous_1['block_id'].values
```

```python
df_anomalous_1 = data_train[data_train['block_id'].isin(block_id_anomalous_1)]
```

```python
df_anomalous_1
```

```python
df_upsample = pd.concat([df_anomalous_1]*10, axis=0)
df_label_upsample = pd.concat([df_label_anomalous_1]*10, axis=0)
```

```python
data_train_upsample = pd.concat([data_train, df_upsample], axis=0)
data_train_label_upsample = pd.concat([label_train, df_label_upsample])
```

```python
len(data_train_upsample), len(data_train_label_upsample)
```

```python
data_train_upsample.to_csv('../analysis/train_upsample.csv', index=False)
data_train_label_upsample.to_csv('../analysis/train_upsample_label.csv', index=False)
```
