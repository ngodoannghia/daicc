```python
import pandas as pd
```

```python
data_train = pd.read_csv('../analysis/val.csv')
label_train = pd.read_csv('../analysis/train_label.csv')
```

```python
num_features = list(data_train.columns)[2:]
```

```python
df_add = data_train.groupby('block_id')[num_features].rank(pct=True).add_prefix('rank_')
```

```python
df_add
```

```python
data_train_pct_rank = pd.concat([data_train, df_add], axis=1)
```

```python
data_train_pct_rank.head()
```

```python
data_train_pct_rank.to_csv('../analysis/pct_rank/val_pct_rank.csv', index=False)
```
